```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./Dataset.csv')
pd.set_option("display.max_columns",None)
```

```python
df.info()
```

```python
df.head(7)
```

```python
df['turning_radius'] = df['turning_radius']*1000
df.turning_radius
```

```python
df.turning_radius.unique()
```

```python
new_list = list()
translation_table = str.maketrans({'"': None, '_': None})
df["population_density"] = [str(item).translate(translation_table) for item in df["population_density"]]
df["population_density"] = df["population_density"].astype(int)
```

```python
tork_in_Nm = []
tork_in_rmp = []
for val in df['max_torque']:
   
    tork_in_Nm.append(val.split("@")[0])
    tork_in_rmp.append(val.split("@")[1])
ser = pd.Series(tork_in_Nm)
tork_in_Nm = ser.str.extract(r'(\d+)', expand=False)
tork_in_rmp = pd.Series(tork_in_rmp).str.extract(r'(\d+)', expand=False)

max_power_in_bhp = []
max_power_in_rpm = []

for val in df["max_power"]:
    max_power_in_bhp.append(val.split("@")[0])
    max_power_in_rpm.append(val.split("@")[1])

max_power_in_bhp = pd.Series(max_power_in_bhp).str.extract(r'(\d+)', expand=False)
max_power_in_rpm = pd.Series(max_power_in_rpm).str.extract(r'(\d+)', expand=False)

```

```python
width_list = []
length_list = []
weight_list = []
```

```python
width_list.append(df['width'].str.split('(',expand=True)[0])
width_list
```

```python

length_list.append(df['Length'].str.rsplit(')',expand=True)[1])
length_list
```

```python
weight_list.append(df['Gross_weight'].str.split('K',expand=True)[0])
weight_list
```

```python
df_len_list = pd.DataFrame(length_list)
df_wei_list = pd.DataFrame(weight_list)
df_wid_list = pd.DataFrame(width_list)
```

```python
ohot_df = pd.get_dummies(df,columns = ['area_cluster','segment','model','fuel_type','is_esc','is_adjustable_steering','is_tpms','is_parking_sensors','is_parking_camera','rear_brakes_type','transmission_type','steering_type','is_front_fog_lights','is_rear_window_wiper','is_rear_window_washer','is_rear_window_defogger', 'is_brake_assist',
       'is_power_door_locks', 'is_central_locking', 'is_power_steering',
       'is_driver_seat_height_adjustable', 'is_day_night_rear_view_mirror',
       'is_ecw', 'is_speed_alert'])

ohot_df = ohot_df.replace({True:1 , False:0})

```

```python
ohot_df.drop(['Length','width','Gross_weight'],axis = 1, inplace = True)
```

```python
ohot_df.drop(['max_torque','engine_type','policy_id'],axis = 1, inplace = True)
```

```python
ohot_df.drop(['max_power'],axis=1, inplace=True)
```

```python
df_len = df_len_list.transpose()
df_wei = df_wei_list.transpose()
df_wid = df_wid_list.transpose()
```

```python
ohot_df['Length'] = df_len
ohot_df['Weight'] = df_wei
ohot_df['Width'] = df_wid

ohot_df['Torque_nm'] = tork_in_Nm
ohot_df['Torque_in_rpm'] = tork_in_rmp
ohot_df['power_in_bhp'] = max_power_in_bhp
ohot_df['power_in_rpm'] = max_power_in_rpm
```

```python
ohot_df['Torque_nm'] = ohot_df['Torque_nm'].astype(int)
ohot_df['Torque_in_rpm'] = ohot_df['Torque_in_rpm'].astype(int)
ohot_df['Length'] = ohot_df['Length'].astype(int)
ohot_df['Weight'] = ohot_df['Weight'].astype(int)
ohot_df['Width'] = ohot_df['Width'].astype(int)
ohot_df['power_in_bhp'] = ohot_df['power_in_bhp'].astype(int)
ohot_df['power_in_rpm'] = ohot_df['power_in_rpm'].astype(int)
```
